```python
class DecisionTreeRegressorCustom:
    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = None if max_depth is None else int(max_depth)
        self.min_samples_split = int(min_samples_split)
        self.min_samples_leaf = int(min_samples_leaf)
        self.tree_ = None

    def _mse(self, y):
        if len(y) == 0:
            return 0.0
        return np.mean((y - np.mean(y)) ** 2)

    def _build(self, X, y, depth=0):
        node = {}
        n_samples = X.shape[0]
        node['n_samples'] = n_samples
        node['value'] = np.mean(y)

        # stopping criteria
        if (self.max_depth is not None and depth >= self.max_depth) or n_samples < self.min_samples_split or n_samples <= 2 * self.min_samples_leaf:
            node['is_leaf'] = True
            return node

        best_feat, best_thr, best_gain = None, None, 0.0
        parent_mse = self._mse(y)

        for j in range(X.shape[1]):
            xj = X[:, j]
            order = np.argsort(xj)
            x_sorted = xj[order]
            y_sorted = y[order]

            uniq = np.unique(x_sorted)
            if uniq.shape[0] == 1:
                continue
            thresholds = (uniq[:-1] + uniq[1:]) / 2.0

            for thr in thresholds:
                left_mask = xj <= thr
                right_mask = ~left_mask
                n_left = np.sum(left_mask)
                n_right = np.sum(right_mask)

                if n_left < self.min_samples_leaf or n_right < self.min_samples_leaf:
                    continue

                mse_left = self._mse(y[left_mask])
                mse_right = self._mse(y[right_mask])
                weighted_mse = (n_left * mse_left + n_right * mse_right) / n_samples
                gain = parent_mse - weighted_mse

                if gain > best_gain:
                    best_gain = gain
                    best_feat = j
                    best_thr = thr

        if best_feat is None or best_gain <= 0:
            node['is_leaf'] = True
            return node

        # split
        left_mask = X[:, best_feat] <= best_thr
        right_mask = ~left_mask
        node['is_leaf'] = False
        node['feature'] = best_feat
        node['threshold'] = best_thr
        node['left'] = self._build(X[left_mask], y[left_mask], depth + 1)
        node['right'] = self._build(X[right_mask], y[right_mask], depth + 1)
        return node

    def fit(self, X, y):
        X_arr = X.values if hasattr(X, 'values') else np.asarray(X)
        y_arr = y.values if hasattr(y, 'values') else np.asarray(y)

        if X_arr.ndim == 1:
            X_arr = X_arr.reshape(-1, 1)

        self.n_features_in_ = X_arr.shape[1]
        self.tree_ = self._build(X_arr.astype(float), y_arr.ravel().astype(float), depth=0)
        return self

    def _predict_row(self, row):
        node = self.tree_
        while not node.get('is_leaf', True):
            if row[node['feature']] <= node['threshold']:
                node = node['left']
            else:
                node = node['right']
        return node.get('value', 0.0)

    def predict(self, X):
        if self.tree_ is None:
            raise RuntimeError("Model has not been fitted yet.")
        
        Xq = X.values if hasattr(X, 'values') else np.asarray(X)
        if Xq.ndim == 1:
            Xq = Xq.reshape(1, -1)
        preds = np.array([self._predict_row(row) for row in Xq], dtype=float)
        return preds
```

```python
class DecisionTreeClassifierCustom:
    def __init__(self, max_depth=None, min_samples_split=2, min_samples_leaf=1):
        self.max_depth = None if max_depth is None else int(max_depth)
        self.min_samples_split = int(min_samples_split)
        self.min_samples_leaf = int(min_samples_leaf)
        self.tree_ = None
        self.classes_ = None

    def _gini(self, y):
        """Вычисляет коэффициент Джини"""
        if len(y) == 0:
            return 0.0
        _, counts = np.unique(y, return_counts=True)
        p = counts / len(y)
        return 1.0 - np.sum(p ** 2)

    def _build(self, X, y, depth=0):
        node = {}
        n_samples = X.shape[0]
        node['n_samples'] = n_samples
        
        # Наиболее частый класс для этого узла
        unique_classes, counts = np.unique(y, return_counts=True)
        node['value'] = unique_classes[np.argmax(counts)]
        node['class_counts'] = dict(zip(unique_classes, counts))

        # Критерий остановки
        if (self.max_depth is not None and depth >= self.max_depth) or \
           n_samples < self.min_samples_split or \
           len(np.unique(y)) == 1:
            node['is_leaf'] = True
            return node

        best_feat, best_thr, best_gain = None, None, 0.0
        parent_gini = self._gini(y)

        # Перебираем признаки
        for j in range(X.shape[1]):
            xj = X[:, j]
            unique_vals = np.unique(xj)
            
            if len(unique_vals) == 1:
                continue
            
            # Генерируем пороги между уникальными значениями
            thresholds = (unique_vals[:-1] + unique_vals[1:]) / 2.0

            for thr in thresholds:
                left_mask = xj <= thr
                right_mask = ~left_mask
                n_left = np.sum(left_mask)
                n_right = np.sum(right_mask)

                if n_left < self.min_samples_leaf or n_right < self.min_samples_leaf:
                    continue

                gini_left = self._gini(y[left_mask])
                gini_right = self._gini(y[right_mask])
                weighted_gini = (n_left * gini_left + n_right * gini_right) / n_samples
                gain = parent_gini - weighted_gini

                if gain > best_gain:
                    best_gain = gain
                    best_feat = j
                    best_thr = thr

        if best_feat is None or best_gain <= 0:
            node['is_leaf'] = True
            return node

        # Разбиваем данные
        left_mask = X[:, best_feat] <= best_thr
        right_mask = ~left_mask
        node['is_leaf'] = False
        node['feature'] = best_feat
        node['threshold'] = best_thr
        node['left'] = self._build(X[left_mask], y[left_mask], depth + 1)
        node['right'] = self._build(X[right_mask], y[right_mask], depth + 1)
        return node

    def fit(self, X, y):
        X_arr = X.values if hasattr(X, 'values') else np.asarray(X)
        y_arr = y.values if hasattr(y, 'values') else np.asarray(y)

        if X_arr.ndim == 1:
            X_arr = X_arr.reshape(-1, 1)

        self.n_features_in_ = X_arr.shape[1]
        self.classes_ = np.unique(y_arr)
        self.tree_ = self._build(X_arr.astype(float), y_arr.ravel().astype(int), depth=0)
        return self

    def _predict_row(self, row):
        node = self.tree_
        while not node.get('is_leaf', True):
            if row[node['feature']] <= node['threshold']:
                node = node['left']
            else:
                node = node['right']
        return node.get('value', self.classes_[0])

    def predict(self, X):
        if self.tree_ is None:
            raise RuntimeError("Model has not been fitted yet.")
        
        Xq = X.values if hasattr(X, 'values') else np.asarray(X)
        if Xq.ndim == 1:
            Xq = Xq.reshape(1, -1)
        preds = np.array([self._predict_row(row) for row in Xq], dtype=int)
        return preds

    def _predict_proba_row(self, row):
        """Возвращает вероятности для каждого класса"""
        node = self.tree_
        while not node.get('is_leaf', True):
            if row[node['feature']] <= node['threshold']:
                node = node['left']
            else:
                node = node['right']
        
        # Получаем counts из листового узла
        class_counts = node.get('class_counts', {})
        total = sum(class_counts.values())
        
        proba = np.zeros(len(self.classes_))
        for i, cls in enumerate(self.classes_):
            proba[i] = class_counts.get(cls, 0) / total if total > 0 else 0.0
        return proba

    def predict_proba(self, X):
        if self.tree_ is None:
            raise RuntimeError("Model has not been fitted yet.")
        
        Xq = X.values if hasattr(X, 'values') else np.asarray(X)
        if Xq.ndim == 1:
            Xq = Xq.reshape(1, -1)
        
        proba_result = np.array([self._predict_proba_row(row) for row in Xq])
        return proba_result
```
